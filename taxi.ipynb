#  Прогнозирование заказов такси

Компания «Чётенькое такси» собрала исторические данные о заказах такси в аэропортах. Чтобы привлекать больше водителей в период пиковой нагрузки, нужно спрогнозировать количество заказов такси на следующий час. Постройте модель для такого предсказания.

Значение метрики *RMSE* на тестовой выборке должно быть не больше 48.

Вам нужно:

1. Загрузить данные и выполнить их ресемплирование по одному часу.
2. Проанализировать данные.
3. Обучить разные модели с различными гиперпараметрами. Сделать тестовую выборку размером 10% от исходных данных.
4. Проверить данные на тестовой выборке и сделать выводы.


Данные лежат в файле `taxi.csv`. Количество заказов находится в столбце `num_orders` (от англ. *number of orders*, «число заказов»).

## Обучение

```python
#начнем с подготовки признаков

def make_features(df, max_lag, rolling_mean_size):
    data_taxi = df.copy()
    data_taxi['year'] = df.index.year
    data_taxi['month'] = df.index.month
    data_taxi['day'] = df.index.day
    data_taxi['dayofweek'] = df.index.dayofweek
    
    for lag in range(1, max_lag + 1):
        data_taxi['lag_{}'.format(lag)] = data_taxi['num_orders'].shift(lag)

    data_taxi['rolling_mean'] = data_taxi['num_orders'].shift().rolling(rolling_mean_size).mean()
    
    return data_taxi
```

<div class="alert alert-block alert-info">
<b>Совет: </b> Желательно было написать функцию таким образом, чтобы она возвращала новый датафрейм, а не изменяла глобальную переменную.
</div>

<div class="alert alert-block alert-warning">
<b>Изменения:</b> Переписала функцию.
</div>

<div class="alert alert-block alert-success">
<b>Успех (ревью 2):</b> Спасибо!
</div>

```python
#Скользящее среднее как признак задаёт общий тренд временного ряда
#Предыдущие значения временного ряда подскажут, будет ли функция x(t) расти или уменьшаться

data_taxi = make_features(data, 48, 5) 
data_taxi.head(7)
```

```python
#разделим данные на обучающую, валидационную и тестовую выборки, размер тестовой - 10% от всех данных

train, test = train_test_split(data_taxi, shuffle=False, test_size=0.1)
train = train.dropna()

#выделим признаки и целевой признак

features_train = train.drop('num_orders', axis=1)
features_test = test.drop('num_orders', axis=1)
target_train = train['num_orders']
target_test = test['num_orders']

#и создадим функцию для подсчета RMSE

def rmse(target, predictions):
    rmse = mean_squared_error(target, predictions, squared=False)
    return rmse
```

```python
scorer = make_scorer(rmse, greater_is_better=False)
tscv = TimeSeriesSplit(n_splits=5) #кросс-валидация для временного ряда
```

### Decision Tree

```python
#найдём наилучшую оценку для модели DecisionTreeRegressor

model_tree = DecisionTreeRegressor()
tree_parametrs = {'max_depth': range (1, 26, 1),
                 'random_state': [12345]}

tree_grid = GridSearchCV(model_tree, tree_parametrs, scoring=scorer, cv=tscv)
tree_grid.fit(features_train, target_train)
print(abs(tree_grid.best_score_))
```

### Random Forest

```python
forest_model = RandomForestRegressor()
parametrs = { 'n_estimators': range (10, 31, 10),
              'max_depth': range (1, 11),
              'random_state': [12345]}
grid = GridSearchCV(forest_model, parametrs, scoring=scorer, cv=tscv)
grid.fit(features_train, target_train)
print(abs(grid.best_score_))
```

### LGBM

```python
lgb = LGBMRegressor()
lgb_parametrs = {'max_depth': range(10, 41, 10),
                 'learning_rate' : [0.5],
                 'n_estimators' : range (5, 16, 5),
                 'random_state': [12345]}

lgb_grid = GridSearchCV(lgb, lgb_parametrs, scoring=scorer, cv=tscv)
lgb_grid.fit(features_train, target_train)
print(abs(lgb_grid.best_score_))
```

### Cat Boost

```python
catboost = CatBoostRegressor() 

cb_parametrs = {'iterations': [100],
                'depth': range(5, 22, 3),
                'learning_rate': [0.5],
               'random_state': [12345]}

cb_grid = GridSearchCV(catboost, cb_parametrs, scoring=scorer, cv=tscv)
cb_grid.fit(features_train, target_train)
print(abs(cb_grid.best_score_))
```
